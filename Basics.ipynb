## Create a python class that takes parameters of Name and Date of Birth and Methods
- a method "setFirstname" that changes the first Name 
- a method  "getFirstname" that returns the Name.
- a method "getAge" that return age of the user. 

```python
class User:
    def __init__(self,firstname,dob):
        self.firstname = firstname
        self.dob = dob
        
    def setFirstname(self,firstname):
        self.firstname = firstname
        return self.firstname 
    def getAge(self):
        from datetime import date
        yr_day = 365
        self.age = int((date.today()-self.dob).days/yr_day)
        return self.age
    def getFirstname (self):
        return self.firstname
```

```python
u2 = User("Asura",date(2012,12,12))
```
