## K-Nearest Neighbors Algorithm
### For different K values i.e, k = 3,5,7

```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_excel(r"C:\Users\sande\OneDrive\Desktop\Practice python\Data Sets\assignment6.xlsx")
```

```python
df
```

```python
# Defining Euclidean distance 
def dis(x,y):
    return np.sqrt(np.sum(x-y)**2)
```

```python
# Given Prediction Point which needs to be assigned by Knn
pred = [8.2,2.6,3.2,1.8]
```

```python
# Creating a new dataset ds for dropping species to calculate distance of all the data to our given prediction point 
ds = df.drop('species',axis =1)
```

```python
ds
```

```python
i = 0
distance = []
while i <=9:
    
    dist = dis(ds.iloc[i],pred)
    i = i+1
    distance.append(dist)
```

```python
distance # Distance from our prediction point
```

```python
df['dist'] = distance
```

```python
df
```

```python
df_nn =  df.sort_values(by=['dist'], axis=0)[:3] # for k = 3
df_nn
```

### Our prediction model is showing it is versicolor for k = 3 

```python
# for k = 5
df_nn =  df.sort_values(by=['dist'], axis=0)[:5]
df_nn 
```

### for k = 5 our predicted point still lies in versicolor 

```python
df_nn =  df.sort_values(by=['dist'], axis=0)[:7] # for k = 7
df_nn # For k = 7 our predicted point is lies in versicolor
```

```python
df_nn.loc[df_nn['species']=='versicolor'].count()>df_nn.loc[df_nn['species']=='setosa'].count()
```

### for k = 7 Hence k will be alloted to versicolor as its closer
