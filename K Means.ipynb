## K Means

```python
df = pd.read_csv(r"C:\Users\sande\OneDrive\Desktop\MMSO\Machine Learning\Categorical.csv")
```

```python
df.head()
```

```python
ds = df.loc[:,["Longitude","Latitude"]]   # Creating a new dataset with only longitude and Latitude
```

```python
ds.head()
```

```python
k = 4   # As we need four clusters  
centeroids = ds.loc[np.random.choice(len(ds['Latitude']),k,replace= False)]  # Taking 4 random rows from dataset ds as centroids 
```

```python
centeroids
```

```python
dis = np.zeros((len(ds["Latitude"]),4))
for i in range (k):
    dis [:,i] = ((ds - centeroids.iloc[i])**2).sum(axis=1)**0.5
dis # Calculating distance(Using Eucilidean) from centeroid
```

```python
close = np.argmin(dis,axis = 1)
```

```python
close # Distributing all the data points to its nearest centroids and distributing them as 0,1,2,3 
```

```python
import matplotlib.colors
custom_cmap = matplotlib.colors.ListedColormap(["Red","Green","Black","Blue"])
plt.figure(figsize=(20, 20))
            
```

```python
plt.scatter(ds['Longitude'],ds['Latitude'],c = close,cmap=custom_cmap)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
